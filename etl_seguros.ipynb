```python
def normalizar_columnas(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
    )
    return df
```

```python
def limpiar_texto_basico(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace(
            ["nan", "None", "NULL", "null", "", "NaN", "N/A", "na"],
            pd.NA
        )
    return df
```

```python
def convertir_fecha_multiple(serie):
    return pd.to_datetime(serie, errors="coerce", dayfirst=False)
```

```python
def estandarizar_genero(genero):
    if pd.isna(genero):
        return pd.NA

    genero = str(genero).strip().lower()

    if genero in ['m', 'masculino', 'male', 'hombre', 'h', 'M']:
        return 'Masculino'
    elif genero in ['f', 'femenino', 'female', 'mujer', 'muj', 'F']:
        return 'Femenino'
    elif genero in ['otro', 'other', 'no binario', 'nb']:
        return 'Otro'
    else:
        return pd.NA
```

```python
#Copia de clientes
clientes_limpio = clientes.copy()
```

```python
#Aplicacmos funciones
clientes_limpio = normalizar_columnas(clientes_limpio)
print("✅ Columnas normalizadas")
clientes_limpio = limpiar_texto_basico(clientes_limpio)
print("✅ Texto limpiado")
if 'fecha_nacimiento' in clientes_limpio.columns:
    clientes_limpio['fecha_nacimiento'] = convertir_fecha_multiple(clientes_limpio['fecha_nacimiento'])
    print("✅ Fecha convertida")

if 'genero' in clientes_limpio.columns:
    clientes_limpio['genero'] = clientes_limpio['genero'].apply(estandarizar_genero)
    print("✅ Género estandarizado")
```

```python
for col in ['nombre', 'ciudad', 'segmento']:
    if col in clientes_limpio.columns:
        mask = clientes_limpio[col].notna()
        clientes_limpio.loc[mask, col] = clientes_limpio.loc[mask, col].str.title()
        print(f"✅ {col} formateado")
```

```python
#Email a minúsculas
clientes_limpio["email"] = clientes_limpio["email"].str.lower()
print("✅ 4.1 Emails convertidos a minúsculas")
```

```python
# 4.2 Mapeo de género
map_genero = {
    "hombre": "Masculino",
    "masculino": "Masculino",
    "m": "Masculino",
    "mujer": "Femenino",
    "femenino": "Femenino",
    "f": "Femenino"
}
```

```python
# Aplicar mapeo solo a valores no nulos
mask_genero = clientes_limpio["genero"].notna()
clientes_limpio.loc[mask_genero, "genero"] = (
    clientes_limpio.loc[mask_genero, "genero"]
    .str.lower()
    .map(map_genero)
)
print("✅ 4.2 Género estandarizado (hombre→Masculino, m→Masculino, mujer→Femenino, etc.)")
```

```python
# 4.3 Mapeo de ciudades
map_ciudad = {
    "sanmiguel": "San Miguel",
    "s. miguel": "San Miguel",
    "san miguel": "San Miguel",
    "sta. ana": "Santa Ana",
    "sta ana": "Santa Ana",
    "santa ana": "Santa Ana",
    "ss": "San Salvador",
    "san salvador": "San Salvador"
}
```

```python
# Aplicar mapeo a ciudades
clientes_limpio["ciudad"] = (
    clientes_limpio["ciudad"]
    .str.lower()
    .replace(map_ciudad)
)
```

```python
# Aplicar title después del mapeo
clientes_limpio["ciudad"] = clientes_limpio["ciudad"].str.title()
print("✅ 4.3 Ciudades normalizadas (ss→San Salvador, sanmiguel→San Miguel, etc.)")
```

```python
# 4.4 Convertir fecha de nacimiento
clientes_limpio["fecha_nacimiento"] = convertir_fecha_multiple(clientes_limpio["fecha_nacimiento"])
print("✅ 4.4 Fecha de nacimiento convertida a datetime")
```

```python
# 4.5 Segmento: llenar nulos con "No especificado"
clientes_limpio["segmento"] = clientes_limpio["segmento"].fillna("No especificado")
print("✅ 4.5 Segmento: nulos reemplazados por 'No especificado'")
```

Resultados de limpieza

```python
# ============================================
# CORRECCIÓN DE PROBLEMAS DETECTADOS
# ============================================

print("\n" + "=" * 60)
print("CORRIGIENDO PROBLEMAS DETECTADOS")
print("=" * 60)

# --------------------------------------------
# 1. CORREGIR CIUDADES INCONSISTENTES
# --------------------------------------------
print("\n📍 Corrigiendo ciudades...")

# Ampliar el mapa de ciudades con más variantes
map_ciudad_ampliado = {
    "sanmiguel": "San Miguel",
    "s. miguel": "San Miguel",
    "san miguel": "San Miguel",
    "s miguel": "San Miguel",
    "sta. ana": "Santa Ana",
    "sta ana": "Santa Ana",
    "santa ana": "Santa Ana",
    "santaana": "Santa Ana",
    "ss": "San Salvador",
    "san salvador": "San Salvador",
    "sansalvador": "San Salvador",
    "sonso": "Sonsonate",
    "sonsonate": "Sonsonate",
    "lalibertad": "La Libertad",
    "la libertad": "La Libertad",
    "l. libertad": "La Libertad"
}

# Aplicar mapeo ampliado
clientes_limpio["ciudad"] = (
    clientes_limpio["ciudad"]
    .str.lower()
    .str.strip()
    .replace(map_ciudad_ampliado)
    .str.title()
)

print("✅ Ciudades corregidas:")
print(clientes_limpio["ciudad"].value_counts().head(10))

# --------------------------------------------
# 2. CORREGIR FECHAS DE NACIMIENTO
# --------------------------------------------
print("\n📍 Corrigiendo fechas de nacimiento...")

# Ver qué fechas son futuras
fechas_futuras = clientes_limpio[clientes_limpio["fecha_nacimiento"] > pd.Timestamp.now()]
print(f"   Fechas futuras encontradas: {len(fechas_futuras)}")

# Establecer fecha límite (personas mayores de 100 años)
fecha_minima = pd.Timestamp.now() - pd.DateOffset(years=100)
fecha_maxima = pd.Timestamp.now()

# Marcar como nulas las fechas fuera de rango
clientes_limpio.loc[
    (clientes_limpio["fecha_nacimiento"] < fecha_minima) |
    (clientes_limpio["fecha_nacimiento"] > fecha_maxima),
    "fecha_nacimiento"
] = pd.NA

print(f"✅ Fechas inválidas marcadas como nulas")
print(f"   Rango válido: {fecha_minima.year} - {fecha_maxima.year}")

# Calcular edad (opcional)
hoy = pd.Timestamp.now()
clientes_limpio["edad"] = (hoy - clientes_limpio["fecha_nacimiento"]).dt.days // 365
print(f"✅ Edad calculada para {clientes_limpio['edad'].notna().sum()} registros")

# --------------------------------------------
# 3. ANÁLISIS DE NULOS EN FECHA
# --------------------------------------------
print("\n Analizando nulos en fecha de nacimiento...")

# Ver si hay patrón en los nulos de fecha
print("\n📊 Distribución de nulos en fecha por segmento:")
print(pd.crosstab(
    clientes_limpio["fecha_nacimiento"].isna(),
    clientes_limpio["segmento"]
))

print("\n📊 Distribución de nulos en fecha por género:")
print(pd.crosstab(
    clientes_limpio["fecha_nacimiento"].isna(),
    clientes_limpio["genero"]
))
```

```python
print("\n" + "-" * 40)
print("5. RESULTADOS DE LIMPIEZA")
print("-" * 40)

print("\n📊 Tipos de datos:")
print(clientes_limpio.dtypes)

print("\n📊 Valores nulos después de limpieza:")
print(clientes_limpio.isnull().sum())

print("\n📊 Distribución de géneros:")
print(clientes_limpio["genero"].value_counts(dropna=False))

print("\n📊 Distribución de ciudades (primeras 10):")
print(clientes_limpio["ciudad"].value_counts().head(10))

print("\n📊 Distribución de segmentos:")
print(clientes_limpio["segmento"].value_counts(dropna=False))

print("\n📊 Rango de fechas de nacimiento:")
print(f"  Mínima: {clientes_limpio['fecha_nacimiento'].min()}")
print(f"  Máxima: {clientes_limpio['fecha_nacimiento'].max()}")
```

```python
# ============================================
# LIMPIEZA DE ASEGURADORAS
# ============================================

print("=" * 60)
print(" LIMPIEZA DE ASEGURADORAS")
print("=" * 60)

# Hacer una copia para trabajar
aseguradoras_limpio = aseguradoras.copy()

# --------------------------------------------
# PASO 1: Normalización básica
# --------------------------------------------
print("\n PASO 1: Normalización básica")

# Normalizar columnas
aseguradoras_limpio.columns = (
    aseguradoras_limpio.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print(" Columnas normalizadas")

# Limpiar texto
for col in aseguradoras_limpio.select_dtypes(include="object").columns:
    aseguradoras_limpio[col] = aseguradoras_limpio[col].astype(str).str.strip()
    aseguradoras_limpio[col] = aseguradoras_limpio[col].replace(
        ["nan", "None", "NULL", "null", ""],
        pd.NA
    )

print(" Texto básico limpiado")

# --------------------------------------------
# PASO 2: Limpieza específica de PAÍS
# --------------------------------------------
print("\n PASO 2: Limpieza de PAÍS")

# Primero, aplicar title()
aseguradoras_limpio["pais"] = aseguradoras_limpio["pais"].str.title()

# Corregir "ElSalvador" a "El Salvador"
aseguradoras_limpio["pais"] = aseguradoras_limpio["pais"].replace(
    "Elsalvador", "El Salvador"
)

# Mostrar resultados
print(" Países normalizados:")
print(aseguradoras_limpio["pais"].value_counts(dropna=False))

# --------------------------------------------
# PASO 3: Limpieza específica de RATING
# --------------------------------------------
print("\n PASO 3: Limpieza de RATING")

# Mapeo + valores adicionales encontrados
map_rating_completo = {
    "alto": "Alto",
    "bajo": "Bajo",
    "medio": "Medio",
    "b": "Bajo",
    "d": "Deficiente",
    "a": "Alto"
}

# Aplicar mapeo
mask_rating = aseguradoras_limpio["rating_riesgo"].notna()
aseguradoras_limpio.loc[mask_rating, "rating_riesgo"] = (
    aseguradoras_limpio.loc[mask_rating, "rating_riesgo"]
    .astype(str)
    .str.lower()
    .map(map_rating_completo)
)

print(" Ratings mapeados:")
print(aseguradoras_limpio["rating_riesgo"].value_counts(dropna=False))

# --------------------------------------------
# PASO 4: Manejar nulos según guía
# --------------------------------------------
print("\n PASO 4: Manejo de nulos")

aseguradoras_limpio["pais"] = aseguradoras_limpio["pais"].fillna("No especificado")
aseguradoras_limpio["rating_riesgo"] = aseguradoras_limpio["rating_riesgo"].fillna("No especificado")

print(" Nulos reemplazados por 'No especificado'")
print(f"\n Valores nulos después de limpieza:")
print(aseguradoras_limpio.isnull().sum())
```

```python
# --------------------------------------------
#  Verificar resultados finales
# --------------------------------------------
print("\n PASO 5: Resultados finales")
print("\n Primeras 5 filas limpias:")
print(aseguradoras_limpio.head())

print("\n Distribución final de países:")
print(aseguradoras_limpio["pais"].value_counts())

print("\n Distribución final de ratings:")
print(aseguradoras_limpio["rating_riesgo"].value_counts())
```

```python
# ============================================
# VALIDACIÓN DE ASEGURADORAS
# ============================================

print("\n" + "=" * 60)
print(" VALIDACIÓN - ASEGURADORAS")
print("=" * 60)

# Reglas de validación:
# - id_aseguradora NO nulo (es clave primaria)
# - nombre NO nulo (campo obligatorio)

print("\n Reglas aplicadas:")
print("   • id_aseguradora NO puede ser nulo")
print("   • nombre NO puede ser nulo")

aseguradoras_validos = aseguradoras_limpio[
    aseguradoras_limpio["id_aseguradora"].notna() &
    aseguradoras_limpio["nombre"].notna()
].copy()

aseguradoras_rechazados = aseguradoras_limpio[
    aseguradoras_limpio["id_aseguradora"].isna() |
    aseguradoras_limpio["nombre"].isna()
].copy()

print(f"\n Total registros: {len(aseguradoras_limpio)}")
print(f" VÁLIDOS: {len(aseguradoras_validos)} ({len(aseguradoras_validos)/len(aseguradoras_limpio)*100:.1f}%)")
print(f" RECHAZADOS: {len(aseguradoras_rechazados)}")

# Agregar motivo de rechazo si hay
if len(aseguradoras_rechazados) > 0:
    def motivo_rechazo(row):
        motivos = []
        if pd.isna(row["id_aseguradora"]):
            motivos.append("id_aseguradora_vacio")
        if pd.isna(row["nombre"]):
            motivos.append("nombre_vacio")
        return ",".join(motivos)

    aseguradoras_rechazados["motivo_rechazo"] = aseguradoras_rechazados.apply(motivo_rechazo, axis=1)
    print("\n Motivos de rechazo:")
    print(aseguradoras_rechazados["motivo_rechazo"].value_counts())
```

```python
# ============================================
# LIMPIEZA DE PÓLIZAS
# ============================================

print("=" * 70)
print("🧹 LIMPIEZA DE PÓLIZAS")
print("=" * 70)

# Hacer una copia para trabajar
polizas_limpio = polizas.copy()
print(f" Total registros: {len(polizas_limpio):,}")

# --------------------------------------------
# PASO 1: Normalización básica
# --------------------------------------------
print("\n PASO 1: Normalización básica")

# Normalizar columnas
polizas_limpio.columns = (
    polizas_limpio.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
print(" Columnas normalizadas")

# Limpiar texto básico (solo columnas object)
for col in polizas_limpio.select_dtypes(include="object").columns:
    polizas_limpio[col] = polizas_limpio[col].astype(str).str.strip()
    polizas_limpio[col] = polizas_limpio[col].replace(
        ["nan", "None", "NULL", "null", ""],
        pd.NA
    )
print(" Texto básico limpiado")

# --------------------------------------------
# PASO 2: Función para limpiar números mixtos
# --------------------------------------------
print("\n PASO 2: Definir función limpiar_numero_mixto")

def limpiar_numero_mixto(valor):
    """
    Limpia números con formatos mixtos:
    - "1.234,56" (europeo) → 1234.56
    - "1,234.56" (americano) → 1234.56
    - "1234.56" → 1234.56
    - "-" → NaN
    """
    if pd.isna(valor):
        return np.nan

    valor = str(valor).strip()

    # Si es guión o vacío, retornar NaN
    if valor in ["", "-", "None", "nan", "NULL", "null"]:
        return np.nan

    # Detectar formato europeo (punto como separador de miles, coma decimal)
    if "," in valor and "." in valor:
        # Si la última coma está después del último punto (formato europeo)
        if valor.rfind(",") > valor.rfind("."):
            # Ejemplo: "27.544,32" → 27544.32
            valor = valor.replace(".", "")  # quitar puntos separadores
            valor = valor.replace(",", ".")  # coma decimal a punto
        else:
            # Ejemplo: "173,298.36" → 173298.36
            valor = valor.replace(",", "")  # quitar comas separadoras
    elif "," in valor:
        # Solo coma: asumimos que es decimal (formato europeo)
        valor = valor.replace(",", ".")

    try:
        return float(valor)
    except:
        return np.nan

print(" Función definida")

# --------------------------------------------
# PASO 3: Función para convertir fechas
# --------------------------------------------
print("\n PASO 3: Definir función convertir_fecha_multiple")

def convertir_fecha_multiple(serie):
    """Convierte fechas en múltiples formatos"""
    return pd.to_datetime(serie, errors="coerce", dayfirst=False)

print(" Función definida")

# --------------------------------------------
# PASO 4: Limpieza específica
# --------------------------------------------
print("\n PASO 4: Aplicando limpieza...")

# 4.1 Fecha de emisión
print("\n Procesando fecha_emision...")
fechas_antes = polizas_limpio['fecha_emision'].isna().sum()
polizas_limpio['fecha_emision'] = convertir_fecha_multiple(polizas_limpio['fecha_emision'])
fechas_despues = polizas_limpio['fecha_emision'].isna().sum()
print(f"   • Fechas nulas: antes {fechas_antes:,} → después {fechas_despues:,}")

# 4.2 Prima
print("\n Procesando prima...")
prima_antes = polizas_limpio['prima'].head(5).tolist()
polizas_limpio['prima'] = polizas_limpio['prima'].apply(limpiar_numero_mixto)
prima_despues = polizas_limpio['prima'].head(5).tolist()
print(f"   • Muestra: {prima_antes} → {prima_despues}")

# 4.3 Comisión
print("\n Procesando comision...")
comision_antes = polizas_limpio['comision'].head(5).tolist()
polizas_limpio['comision'] = polizas_limpio['comision'].apply(limpiar_numero_mixto)
comision_despues = polizas_limpio['comision'].head(5).tolist()
print(f"   • Muestra: {comision_antes} → {comision_despues}")

# 4.4 Monto asegurado
print("\n Procesando monto_asegurado...")
monto_antes = polizas_limpio['monto_asegurado'].head(5).tolist()
polizas_limpio['monto_asegurado'] = polizas_limpio['monto_asegurado'].apply(limpiar_numero_mixto)
monto_despues = polizas_limpio['monto_asegurado'].head(5).tolist()
print(f"   • Muestra: {monto_antes} → {monto_despues}")
```

```python
# ============================================
# LIMPIEZA DE SINIESTROS
# ============================================

print("=" * 70)
print(" LIMPIEZA DE SINIESTROS (ÚLTIMO ARCHIVO)")
print("=" * 70)

# Hacer una copia para trabajar
siniestros_limpio = siniestros.copy()
print(f" Total registros: {len(siniestros_limpio):,}")

# --------------------------------------------
# PASO 1: Normalización básica
# --------------------------------------------
print("\n PASO 1: Normalización básica")

# Normalizar columnas
siniestros_limpio.columns = (
    siniestros_limpio.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
print(" Columnas normalizadas")

# Limpiar texto básico
for col in siniestros_limpio.select_dtypes(include="object").columns:
    siniestros_limpio[col] = siniestros_limpio[col].astype(str).str.strip()
    siniestros_limpio[col] = siniestros_limpio[col].replace(
        ["nan", "None", "NULL", "null", ""],
        pd.NA
    )
print(" Texto básico limpiado")

# --------------------------------------------
# PASO 2: Función para limpiar números mixtos
# --------------------------------------------
print("\n PASO 2: Definir función limpiar_numero_mixto")

def limpiar_numero_mixto(valor):
    """
    Limpia números con formatos mixtos:
    - "1.619,38" (europeo) → 1619.38
    - "7,076.25" (americano) → 7076.25
    - "2092.59" → 2092.59
    """
    if pd.isna(valor):
        return np.nan

    valor = str(valor).strip()

    if valor in ["", "-", "None", "nan", "NULL", "null"]:
        return np.nan

    # Detectar formato europeo (punto como separador de miles, coma decimal)
    if "," in valor and "." in valor:
        # Si la última coma está después del último punto (formato europeo)
        if valor.rfind(",") > valor.rfind("."):
            # Ejemplo: "1.619,38" → 1619.38
            valor = valor.replace(".", "")  # quitar puntos separadores
            valor = valor.replace(",", ".")  # coma decimal a punto
        else:
            # Ejemplo: "7,076.25" → 7076.25
            valor = valor.replace(",", "")  # quitar comas separadoras
    elif "," in valor:
        # Solo coma: asumimos que es decimal (formato europeo)
        valor = valor.replace(",", ".")

    try:
        return float(valor)
    except:
        return np.nan

print(" Función definida")

# --------------------------------------------
# PASO 3: Función para convertir fechas
# --------------------------------------------
print("\n PASO 3: Definir función convertir_fecha_multiple")

def convertir_fecha_multiple(serie):
    """Convierte fechas en múltiples formatos (dd-mm-yyyy, dd/mm/yyyy, yyyy/mm/dd)"""
    return pd.to_datetime(serie, errors="coerce", dayfirst=True)  # dayfirst=True para formatos europeos

print(" Función definida")

# --------------------------------------------
# PASO 4: Limpieza específica
# --------------------------------------------
print("\n PASO 4: Aplicando limpieza...")

# 4.1 Fecha de siniestro
print("\n Procesando fecha_siniestro...")
fechas_antes = siniestros_limpio['fecha_siniestro'].isna().sum()
siniestros_limpio['fecha_siniestro'] = convertir_fecha_multiple(siniestros_limpio['fecha_siniestro'])
fechas_despues = siniestros_limpio['fecha_siniestro'].isna().sum()
print(f"   • Fechas inválidas convertidas a NaN: {fechas_despues - fechas_antes}")

# 4.2 Monto siniestro
print("\n Procesando monto_siniestro...")
print(f"   • Muestra antes: {siniestros_limpio['monto_siniestro'].head(5).tolist()}")
siniestros_limpio['monto_siniestro'] = siniestros_limpio['monto_siniestro'].apply(limpiar_numero_mixto)
print(f"   • Muestra después: {siniestros_limpio['monto_siniestro'].head(5).tolist()}")

# 4.3 Estado
print("\n Procesando estado...")

# Mapeo  + valores encontrados
map_estado = {
    "abierto": "Abierto",
    "abierto": "Abierto",  # para ABIERTO
    "cerrado": "Cerrado",
    "cerrado": "Cerrado",
    "rechazado": "Rechazado",
    # "en proceso" no aparece en los datos
}

print(f"   • Valores únicos antes:")
print(f"      {siniestros_limpio['estado'].value_counts(dropna=False).to_dict()}")

# Aplicar mapeo
mask_estado = siniestros_limpio['estado'].notna()
siniestros_limpio.loc[mask_estado, 'estado'] = (
    siniestros_limpio.loc[mask_estado, 'estado']
    .str.strip()
    .str.lower()
    .map(map_estado)
)

# Llenar nulos con "No especificado"
siniestros_limpio['estado'] = siniestros_limpio['estado'].fillna("No especificado")

print(f"\n   • Valores únicos después:")
print(f"      {siniestros_limpio['estado'].value_counts().to_dict()}")

# --------------------------------------------
# PASO 5: Verificar resultados
# --------------------------------------------
print("\n📌 PASO 5: Resultados finales")
print("\n Valores nulos después de limpieza:")
nulos = siniestros_limpio.isnull().sum()
print(nulos)

print("\n Estadísticas de MONTO_SINIESTRO:")
print(siniestros_limpio['monto_siniestro'].describe())

print("\n Rango de FECHAS:")
print(f"   • Mínima: {siniestros_limpio['fecha_siniestro'].min()}")
print(f"   • Máxima: {siniestros_limpio['fecha_siniestro'].max()}")

print("\n Primeras 5 filas limpias:")
print(siniestros_limpio.head())
```

```python
# ============================================
# CORRECCIÓN: CONVERTIR TIPOS DE DATOS
# ============================================

from sqlalchemy import create_engine, text
import pandas as pd
import numpy as np

# Tu conexión
database_url = "postgresql://etl_seguros_nu8y_user:MhUvzgn0RNlRIpHCo2CTQEZI8PhH2Vpy@dpg-d6qu6tv5gffc73f0dlng-a.oregon-postgres.render.com/etl_seguros_nu8y"
engine = create_engine(database_url)

print("=" * 70)
print("🔄 CONVIRTIENDO TIPOS DE DATOS PARA POSTGRESQL")
print("=" * 70)

# --------------------------------------------
# 1. RECARGAR ARCHIVOS DESDE GITHUB
# --------------------------------------------
print("\n📦 RECARGANDO ARCHIVOS DESDE GITHUB...")

urls = {
    'clientes': 'https://raw.githubusercontent.com/davidromero2534582021/etl-seguros-pipeline/refs/heads/main/data/curated/clientes_curated.csv',
    'aseguradoras': 'https://raw.githubusercontent.com/davidromero2534582021/etl-seguros-pipeline/refs/heads/main/data/curated/aseguradoras_curated.csv',
    'corredores': 'https://raw.githubusercontent.com/davidromero2534582021/etl-seguros-pipeline/refs/heads/main/data/curated/corredores_curated.csv',
    'tipos_seguro': 'https://raw.githubusercontent.com/davidromero2534582021/etl-seguros-pipeline/refs/heads/main/data/curated/tipos_seguro_curated.csv',
    'polizas': 'https://raw.githubusercontent.com/davidromero2534582021/etl-seguros-pipeline/refs/heads/main/data/curated/polizas_curated.csv',
    'siniestros': 'https://raw.githubusercontent.com/davidromero2534582021/etl-seguros-pipeline/refs/heads/main/data/curated/siniestros_curated.csv'
}

dataframes = {}
for nombre, url in urls.items():
    print(f"   • {nombre}: ", end="")
    try:
        df = pd.read_csv(url)
        dataframes[nombre] = df
        print(f"✅ {len(df):,} registros")
    except Exception as e:
        print(f"❌ Error: {e}")

# --------------------------------------------
# 2. FUNCIONES DE CONVERSIÓN
# --------------------------------------------
def convertir_fecha_para_postgres(serie):
    """Convierte fechas al formato ISO (YYYY-MM-DD) que PostgreSQL entiende"""
    if serie is None or pd.isna(serie):
        return None

    try:
        # Intentar convertir con pandas
        fecha = pd.to_datetime(serie, errors='coerce')
        if pd.notna(fecha):
            return fecha.strftime('%Y-%m-%d')
        return None
    except:
        return None

def convertir_numero_para_postgres(valor):
    """Convierte números con formato mixto a float"""
    if valor is None or pd.isna(valor) or valor == '':
        return None

    valor_str = str(valor).strip()

    # Si es guión o texto, retornar None
    if valor_str in ['-', 'None', 'nan', 'NULL', 'null', '']:
        return None

    try:
        # Manejar formatos con coma y punto
        if ',' in valor_str and '.' in valor_str:
            # Formato europeo: 1.234,56
            if valor_str.rfind(',') > valor_str.rfind('.'):
                valor_str = valor_str.replace('.', '')
                valor_str = valor_str.replace(',', '.')
            # Formato americano: 1,234.56
            else:
                valor_str = valor_str.replace(',', '')
        elif ',' in valor_str:
            valor_str = valor_str.replace(',', '.')

        return float(valor_str)
    except:
        return None

# --------------------------------------------
# 3. CONVERTIR CLIENTES
# --------------------------------------------
print("\n📌 CONVIRTIENDO CLIENTES...")
clientes_convertido = dataframes['clientes'].copy()
clientes_convertido['fecha_nacimiento'] = clientes_convertido['fecha_nacimiento'].apply(convertir_fecha_para_postgres)
clientes_convertido['genero'] = clientes_convertido['genero'].replace({'Hombre': 'Masculino', 'Mujer': 'Femenino', 'M': 'Masculino', 'F': 'Femenino'})
print(f"   • Fechas convertidas: {clientes_convertido['fecha_nacimiento'].notna().sum()} válidas")

# --------------------------------------------
# 4. CONVERTIR TIPOS_SEGURO
# --------------------------------------------
print("\n📌 CONVIRTIENDO TIPOS_SEGURO...")
tipos_convertido = dataframes['tipos_seguro'].copy()
tipos_convertido['riesgo_base'] = tipos_convertido['riesgo_base'].apply(convertir_numero_para_postgres)
tipos_convertido['categoria'] = tipos_convertido['categoria'].str.title()
print(f"   • Riesgos convertidos: {tipos_convertido['riesgo_base'].notna().sum()} válidos")

# --------------------------------------------
# 5. CONVERTIR POLIZAS
# --------------------------------------------
print("\n📌 CONVIRTIENDO POLIZAS (puede tomar unos segundos)...")
polizas_convertido = dataframes['polizas'].copy()
polizas_convertido['fecha_emision'] = polizas_convertido['fecha_emision'].apply(convertir_fecha_para_postgres)
polizas_convertido['prima'] = polizas_convertido['prima'].apply(convertir_numero_para_postgres)
polizas_convertido['comision'] = polizas_convertido['comision'].apply(convertir_numero_para_postgres)
polizas_convertido['monto_asegurado'] = polizas_convertido['monto_asegurado'].apply(convertir_numero_para_postgres)
print(f"   • Fechas: {polizas_convertido['fecha_emision'].notna().sum()} válidas")
print(f"   • Primas: {polizas_convertido['prima'].notna().sum()} válidas")

# --------------------------------------------
# 6. CONVERTIR SINIESTROS
# --------------------------------------------
print("\n📌 CONVIRTIENDO SINIESTROS...")
siniestros_convertido = dataframes['siniestros'].copy()
siniestros_convertido['fecha_siniestro'] = siniestros_convertido['fecha_siniestro'].apply(convertir_fecha_para_postgres)
siniestros_convertido['monto_siniestro'] = siniestros_convertido['monto_siniestro'].apply(convertir_numero_para_postgres)
siniestros_convertido['estado'] = siniestros_convertido['estado'].str.title()
print(f"   • Fechas: {siniestros_convertido['fecha_siniestro'].notna().sum()} válidas")
print(f"   • Montos: {siniestros_convertido['monto_siniestro'].notna().sum()} válidos")

# --------------------------------------------
# 7. RECREAR TABLAS Y CARGAR DATOS
# --------------------------------------------
print("\n" + "=" * 70)
print("🏗️ RECREANDO TABLAS Y CARGANDO DATOS")
print("=" * 70)

# Drop y create tablas
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS siniestros CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS polizas CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS tipos_seguro CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS corredores CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS aseguradoras CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS clientes CASCADE"))
    conn.commit()
print("✅ Tablas eliminadas")

# Crear tablas nuevamente
create_tables_sql = """
CREATE TABLE clientes (
    id_cliente INT PRIMARY KEY,
    nombre VARCHAR(150),
    email VARCHAR(150),
    genero VARCHAR(20),
    fecha_nacimiento DATE,
    ciudad VARCHAR(100),
    segmento VARCHAR(50)
);

CREATE TABLE aseguradoras (
    id_aseguradora INT PRIMARY KEY,
    nombre VARCHAR(150),
    pais VARCHAR(100),
    rating_riesgo VARCHAR(50)
);

CREATE TABLE corredores (
    id_corredor INT PRIMARY KEY,
    nombre VARCHAR(150),
    zona VARCHAR(100),
    nivel VARCHAR(50),
    anios_experiencia INT
);

CREATE TABLE tipos_seguro (
    id_tipo_seguro INT PRIMARY KEY,
    tipo VARCHAR(100),
    categoria VARCHAR(100),
    riesgo_base NUMERIC
);

CREATE TABLE polizas (
    id_poliza INT PRIMARY KEY,
    fecha_emision DATE,
    id_cliente INT,
    id_corredor INT,
    id_aseguradora INT,
    id_tipo_seguro INT,
    prima NUMERIC,
    comision NUMERIC,
    monto_asegurado NUMERIC
);

CREATE TABLE siniestros (
    id_siniestro INT PRIMARY KEY,
    id_poliza INT,
    fecha_siniestro DATE,
    monto_siniestro NUMERIC,
    estado VARCHAR(50)
);
"""

with engine.connect() as conn:
    for comando in create_tables_sql.split(';'):
        if comando.strip():
            conn.execute(text(comando))
    conn.commit()
print("✅ Tablas recreadas")

# --------------------------------------------
# 8. CARGAR DATOS CONVERTIDOS
# --------------------------------------------
print("\n📤 CARGANDO DATOS CONVERTIDOS...")

# Diccionario con DataFrames convertidos
datos_convertidos = {
    'clientes': clientes_convertido,
    'aseguradoras': dataframes['aseguradoras'],  # ya está bien
    'corredores': dataframes['corredores'],      # ya está bien
    'tipos_seguro': tipos_convertido,
    'polizas': polizas_convertido,
    'siniestros': siniestros_convertido
}

# Orden de carga
orden_carga = ['clientes', 'aseguradoras', 'corredores', 'tipos_seguro', 'polizas', 'siniestros']

for tabla in orden_carga:
    df = datos_convertidos[tabla]
    print(f"\n📌 Cargando {tabla}... ({len(df):,} registros)")

    try:
        df.to_sql(tabla, engine, if_exists='append', index=False)
        print(f"   ✅ {len(df):,} registros cargados")
    except Exception as e:
        print(f"   ❌ Error: {e}")
        print(f"   Primeros 3 registros:")
        print(df.head(3))

# --------------------------------------------
# 9. VERIFICAR CARGAS
# --------------------------------------------
print("\n" + "=" * 70)
print("🔍 VERIFICANDO CARGAS FINALES")
print("=" * 70)

conteos = pd.read_sql("""
    SELECT 'clientes' as tabla, COUNT(*) FROM clientes
    UNION ALL SELECT 'aseguradoras', COUNT(*) FROM aseguradoras
    UNION ALL SELECT 'corredores', COUNT(*) FROM corredores
    UNION ALL SELECT 'tipos_seguro', COUNT(*) FROM tipos_seguro
    UNION ALL SELECT 'polizas', COUNT(*) FROM polizas
    UNION ALL SELECT 'siniestros', COUNT(*) FROM siniestros
    ORDER BY tabla;
""", engine)

print("\n📊 REGISTROS FINALES POR TABLA:")
print(conteos.to_string(index=False, header=False))

print("\n" + "=" * 70)
print("🎉 ¡CARGA EXITOSA A POSTGRESQL!")
print("=" * 70)
```
